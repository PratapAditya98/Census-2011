```python
#Checking whether the data adds up! Data Validation

def check_india_population_consistency(df):
    # Get India's total population (where Name is 'India' and TRU is 'Total')
    india_total = df.loc[(df['Name'].str.upper() == 'INDIA') & (df['TRU'] == 'Total'), 'TOT_P'].sum()

    # Get sum of total population for all rows where Name != 'India' and TRU == 'Total'
    state_ut_total = df.loc[(df['Name'].str.upper() != 'INDIA') & (df['TRU'] == 'Total'), 'TOT_P'].sum()

    # Compare and return results
    is_equal = india_total == state_ut_total
    difference = india_total - state_ut_total

    return {
        'india_total': india_total,
        'sum_of_states_uts': state_ut_total,
        'matches': is_equal,
        'difference': difference
    }
```
